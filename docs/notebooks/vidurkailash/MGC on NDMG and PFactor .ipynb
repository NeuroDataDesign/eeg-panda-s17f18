# Converting files from gpickle to edgelist 

```python
import os 
import glob
import numpy as np 
import networkx as nx
```

```python
def open_graphs(grabstr, extension='.gpickle'):
    names = os.path.join(*[grabstr + extension])
    fnames = glob.glob(names)
    basenames = [os.path.splitext(fname)[0] for fname in fnames]
    if (extension == '.gpickle'):
        graphs = [nx.read_gpickle(fname) for fname in fnames]
    return (graphs, basenames)
```

```python
def save_graphs(graphs, basenames, extension='.edgelist'):
    for (graph, basename) in zip(graphs, basenames):
        newfile = basename + extension
        if (extension == '.edgelist'):
            nx.write_weighted_edgelist(graph, newfile)
    return
```

```python
def convert_dataset(basepath, dataset, atlases):
    for atlas in atlases:
        print(atlas)
        gpath = basepath.format(dataset, atlas)                  
        (graphs, basenames) = open_graphs(gpath)
        save_graphs(graphs, basenames)
```

```python
graphs1, basenames1 = open_graphs('sub-NDARHP176DPE_task-rest_bold_desikan_res-2x2x2_measure-correlation', extension='.gpickle')
```

```python
save_graphs(graphs1, basenames1, extension='.edgelist')
```

```python
graphs, basenames = open_graphs('sub-NDARAR025WX4_task-rest_bold_desikan_res-2x2x2_measure-correlation', extension='.gpickle')
```
