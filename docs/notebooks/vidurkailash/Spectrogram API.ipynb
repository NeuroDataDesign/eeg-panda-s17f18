```python
import numpy as np 
import pandas as pd
import pickle
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly import tools

init_notebook_mode(connected=True)

print __version__ 
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
def spectrograms(D, p_global, p_local):
    F_s = p_global['sample_freq']
    Patient_ID = p_global['patient_id']
    Title = p_local['paradigm']
    Data = D[Title]
    
    
    disp_chans = range(Data.shape[0])
    if 'disp_chans' in p_local.keys():
        disp_chans = p_local['disp_chans']
    
    
    F_s = 500 # sampling freq
    dt = 1./F_s

    for i in disp_chans:
        sample_points = np.arange(Data.shape[1]) * dt
        signal = Data[i, :]
        ft = np.fft.fft(signal) * dt
        ft = ft[: len(sample_points)/2]
        freq = np.fft.fftfreq(len(sample_points), dt)
        freq = freq[:len(sample_points)/2]

        trace = go.Scatter(
            x = freq[:2000],
            y = np.abs(ft)[:2000],
            mode = 'markers'
        )

        data = [trace]

        layout = go.Layout(
            title = 'Frequency Spectrum, Channel ' + str(i), 
            hovermode = 'closest', 
            xaxis = dict(
                title = 'Hz',
                range = [0, 10]
                ), 
            yaxis = dict(
                title = 'Magnitude',
                range = [0, 1000]
                )
        )

        fig= go.Figure(data=data, layout=layout)
        iplot(fig)
```
