```python
import numpy as np 
import pandas as pd
import pickle
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly import tools

init_notebook_mode(connected=True)

print __version__ 
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
def spectrograms(D, p_global, p_local):
    F_s = p_global['sample_freq']
    Patient_ID = p_global['patient_id']
    Title = p_local['paradigm']
    Data = D[Title]
    
    
    disp_chans = range(Data.shape[0])
    if 'disp_chans' in p_local.keys():
        disp_chans = p_local['disp_chans']
    
    
    F_s = 500 # sampling freq
    dt = 1./F_s

    for i in disp_chans:
        sample_points = np.arange(Data.shape[1]) * dt
        signal = Data[i, :]
        ft = np.fft.fft(signal) * dt
        ft = ft[: len(sample_points)/2]
        freq = np.fft.fftfreq(len(sample_points), dt)
        freq = freq[:len(sample_points)/2]

        
        data = [dict(
            visible = False,
            line=dict(color='00CED1', width=6),
#             name = 'ùúà = '+str(step),
            x = freq[:2000],
            y = np.abs(ft)[:2000],
            mode = 'markers')]
    
    xaxis = dict(
        title = 'Time  seconds timestamps)'
    )
    yaxis = dict(
        title = 'Intensity (microVolts)'
    )

    layout = dict(xaxis=xaxis, yaxis=yaxis)
    
    fig= go.Figure(data=data, layout=layout)
    iplot(fig)
#         steps = []
#     for i in range(len(data)):
#         step = dict(
#             method = 'restyle',
#             args = ['visible', [False] * len(data)],
#         )
#         step['args'][1][i] = True # Toggle i'th trace to "visible"
#         steps.append(step)

#     sliders = [dict(
#         active = 10,
#         currentvalue = {"prefix": "Electrode: "},
#         pad = {"t": 50},
#         steps = steps
#     )]

#     layout = dict(sliders=sliders)

#     fig = dict(data=data, layout=layout)
        

        
#     fig= go.Figure(data=data, layout=layout)
#     iplot(fig)
    
    
```
