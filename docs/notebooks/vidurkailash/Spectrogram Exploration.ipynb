```python
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
F_s = 500 # sampling freq
dt = 1./F_s

fig = tools.make_subplots(rows=2, cols=5)

for i in range(10):
    sample_points = np.arange(NDARAA075AMK['video1'].shape[1]) * dt
    signal = NDARAA075AMK['video1'][i, :]
    ft = np.fft.fft(signal) * dt
    ft = ft[: len(sample_points)/2]
    freq = np.fft.fftfreq(len(sample_points), dt)
    freq = freq[:len(sample_points)/2]

    trace = go.Scatter(
        x = freq[:2000],
        y = np.abs(ft)[:2000],
        mode = 'markers'
    )
    
    fig.append_trace(trace, i / 5 + 1, i % 5 + 1)
    
#     data = [trace]
    
# layout = go.Layout(
#     title = 'Frequency Spectrum, All Channels', 
#     hovermode = 'closest', 
#     xaxis = dict(
#         title = 'Hz',
#         range = [0, 10]
#         ), 
#     yaxis = dict(
#         title = 'Magnitude',
#         range = [0, 1000]
#         )
# )

    fig['layout'].update(
        title = 'Frequency Spectrum, All Channels', 
    #     hovermode = 'closest', 
        xaxis = dict(
            title = 'Hz',
            range = [0, 10]
            ), 
        yaxis = dict(
            title = 'Magnitude',
            range = [0, 1000]
            )
    )
iplot(fig)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import Range1d
from bokeh.models import Title
```

```python
F_s = 500 # sampling freq
dt = 1./F_s

for i in range(11):
    sample_points = np.arange(NDARAA075AMK['video1'].shape[1]) * dt
    signal = NDARAA075AMK['video1'][i, :]
    ft = np.fft.fft(signal) * dt
    ft = ft[: len(sample_points)/2]
    freq = np.fft.fftfreq(len(sample_points), dt)
    freq = freq[:len(sample_points)/2]
    
    x = freq[:2000];
    y = np.abs(ft)[:2000];
    
    p = figure(title= 'Frequency Spectrum, Channel ' + str(i), plot_width=1000, x_range=(0,10))
    p.y_range= Range1d(0,1000)
    p.circle(x, y, size=5)
    p.add_layout(Title(text='Magnitude', align='center'), 'left')
    p.add_layout(Title(text='Hz', align='center'), 'below')
    
    print(i)
    show(p)
```
