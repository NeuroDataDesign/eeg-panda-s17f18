```python
matplotlib.use('agg')
import matplotlib.pyplot as plt
from numpy import genfromtxt
import numpy as np
import seaborn as sns


sns.set(style="white")

first_obs_no = 0
last_obs_no = 100000
# timestep set at 50
sep = 50
columns = []
for i in range(first_obs_no,last_obs_no,sep):
    columns.append(i)


my_data = genfromtxt('NDARAD481FXF_RestingState_data.csv', delimiter=',',usecols=columns)

a = np.corrcoef(my_data)
# get ride of the last row
a = a[:128,:128]
sns.heatmap(a,xticklabels=10, yticklabels=10)
```

```python
my_data = genfromtxt('NDARAD481FXF_RestingState_data.csv', delimiter=',',usecols=columns)
b = np.cov(my_data)
sv = np.linalg.eigvals(b)
sv = sv[:128]
cs = []
col = []
cumsum = 0
sum_sv = sum(sv)

for i in range(len(sv)):
    cumsum = cumsum + (sv[i]/sum_sv) * 100
    # print (sv[i]/sum_sv)
    cs.append(cumsum)
plt.plot(np.r_[0:128],cs)
```
