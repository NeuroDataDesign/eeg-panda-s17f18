# In search of biomarkers in psychiatry


# Importance
- Diagnoses based on evalution of behavior and clinical phenotypes has led to somewhat arbitrary cut-offs for disorder and subtype classifications
- Non-invasive biomarker research in psychiatry has focused largely on neuroimaging as a tool for identifying neural functions that are meaningfully associated with psychopathology
- heterogeneity exists within disovers both at the levels of genetic and environmental risk factors in brain systems
     - Large samples will enable improved understanding and better stratification of the brain functions that underlie disorder phenotypes and their proposed subtypes

The five candidate domains of RDoc are (1) Negative and Positive Valence Systems, (2) Cognitive Systems, (3) Systems for Social Processes and (4) Arousal and (5) Regulatory Processes. These domains and circuits transcend DSM diagnoses to attain further connection between biological abnormalities and symptoms.

As major advances in computer hardware and signal processing greatly increased the amount of spatially and temporally precise information about brain function that can be extracted from EEG data, EEG has become a good neuroimaging technique for search of biomarkers.

ERP attempts to reduce the complexity of the recorded EEG by 
- averaging across epochs time-locked to sets of events assumed to elicit similar brain function
- comparing at most a few resulting channel ERP features to behavioral differences across groups, conditions, and/or treatments.

It is a good approach to identify coherent ERP features with signal-to-noise ratio adequate for stably comparing group, task, and/or treatment conditions. However, there is also a great reduction in the amount of infotmation about brain function extracted from the data.



# Indepedent Component Analysis (ICA)
- ICA decomposition can be used to identify distinct cortical source activities that are sensitive and specific to the pathophysiology of psychiatric disorders
- essentially they learn, from the data itself, spatial filters that separate the EEG data into a sum of component processes with maximally distinct or independent time courses
    - improved understanding of a wide variety of functions in the general population, including face processing, cognitive control and mirror neuron activity
    - has the potential to provide objective markers and phenotypes of the degree of affective and emotional abnormality in psychiatric disorders such as schizophrenia and depression
- ICA is able to do the seprations without priori knowledge of individual properties or locations
- ICA decomposition further increases the signal-to-noise ratio of brain component process measures relative to scalp channel measures
- ICA project to the scalp in a dipolar spatial pattern compatible with the projection of field activity from a single cortical location
    - The best-fitting tiny battery position and orientation together define the equivalent dipole as an approximate source location
    - The spatial locations of both the equivalent dipoles, the actual source locations for independent component processes with dipolar projections can be estimated using an electrical head model, optimally one built from an individual participant magnetic resonance head image though a best-fit standard template head model provides a small margin of localization error


# EEG
EEG has attracted increased interest for development of biomarkers for psychiatric diagnosis and phenotype definition
- advantage of EEG:
    - ease of use
    - non-invasive
    - dry-wireless, wearable technology 
    - lack of strict head movement constriant -- great for young children under critical brain developments
    
- In conclusion:
    - ongoing advances in signal processing and visualization of EEG activity greatly improve the spatial resolution of EEG imaging
    - the relatively low cost of EEG data collection means it is accessible for studies on the large number of samples
    - EEG is the most non-invasive and portable of all the neuroimaging methods
        
- traditionally analyzed in two ways:
    - averages data across many segments that are time-locked to events of interest (only a small percentage of the total recorded EEG signal is reflected)
    - the ongoing EEG channel signals may be processed using spectral analysis, across a time period of interest, to identify the mean or relative power measures for one or more EEG frequency bands of interest (qEEG)
    
    
- large samples in twin or family study have indicated that estimates of the role of genetic factors in EEG measures parallel those found in twin and family studies of behavior and and brain structure and surpass those for fMRI, which was believed to be a more reliable neuroimaging technique for spatial resolution
    - a metaanalysis of twin studies of EEG alpha frequency power indicating a meta-heritability estimate of 80%


- **Biomarkers**, or endophenotypes, are an alternative strategy to more directly assay the effects of disorder risk variants and thus accelerate identification
- p300
    - a positively inflected ERP peak that occur around 300 milliseconds after a task target stimulus
    - related to disorders include alcohol abuse and schizophrenia
- p50
    - reliably associated with schizophrenia
- ERP components related to cognitive control and response inhibition have shown consistent association with attention deficit hyperactivity disorder (ADHD)


- **cocktail party problem**
    - separation of the signal contributions of non-brain sources to the scalp data can be viewed as a blind source separation problem
    - in analogy, think about each microphone channel signal will sum the voices of all the participants with relative strengths that vary according to their distance and spatial arrangement
    - with some statistical assumptions we could apply ICA to “invert” the linear mixing process thereby extracting the individual voice signals from the recorded mixtures
    - "voices"-> spatially coherent local field potential signals (effective source signals)
        - emerging within the vast complexity of cortical electrical dynamics
        - project to the scalp by simple volume conduction
        - linearly mixed at the recording electrodes, with each other and with potentials conducted from non-brain sources(eyes, muscles, heart, etc.)

# Incorporating priors for EEG source imaging and connectivity analysis

EEG has been recognized as a useful technique to non-invasive study of brain dynamics
- easy to manipulate
- cheap to set up
- take real-time responses from neural activity without delay
- high temporal resolution to investigate dynamic brain activation during various cognitive tasks
- great improvements in spatial resolution by increasing the number of electrodes
- currently, priori assumptions need to be incorporated to reduce the number of unknown parameters in order to solve the problem. Available priors in ESI are from MRI, fMRI, PET, DTI and TMS.
- prioris from other modalities are used to validate the correctness of the localization

**Penalty Function**

employs different types of norm and weight matrices to incorporate priors from other modalities
- identity matrix produces the classical minimum norm model (MNM)
    - gain matrix introduces an intrinsic bias that favors the sensor in MNM
- advantages:
    -  both the types of norm and the weight matrix are configurable
        - the choices are valuable since it is easy to design an experiment with specific penalized model
    - achieve balance by combining with multiple penalty functions in a single objective function
    
**Bayesian Framework**

allows a priori assumption of EEG source to be explicitly quantified by using postulated prior distributions
- include info on neural current
- combined spatial and temporal constraints
- sparase nature of sources
- parametric empirical Bayesian(PEB) -> show reliable estimation of EEG sources
- advantages:
    - hierarchical strcture
        - lower level has direct and strong influence on the solution
        - higher level only affect its adjacent levels
            - weak influence on final solution
- can be translate into a quadratic penalty function through transformation of logarithm
    - if the prior distribution satisfies multivariable Gaussian distribution
    
    

# A CONNECTOME WIDE FUNCTIONAL SIGNATURE OF BROAD RISK FOR MENTAL ILLNESS

Aim to find a single general factor 'p'
- associated with risk for all common forms of mental illness
- built on past researches
- use a data-driven analysis of connectome wide intrinsic functional connectivity

‘p’ factor provides a framework for explaining the high rates of comorbidity as well as the shared genetic variance among categorical mental disorders
- a recent large meta-analysis of six categorical disorders reported a shared pattern of **reduced gray matter volume** in a distributed network supporting attention and cognitive control
-  claim of the paper -> **higher p factor scores and thus risk for common mental illness was associated with reduced gray matter volumes in the occipital lobes and neocerebellum**
- higher p factor scores were associated with reduced fractional anisotropy in pontine pathways linking the neocerebellum with the thalamus and prefrontal cortex
- higher p factor scores contribute to broad risk for common forms of mental illness through alterations in networks critical for feed-forward monitoring of information processing and executive control


Resting-state functional connectivity is a poweful tool
- can be readily administered across patient populations
- represents a powerful probe of the intrinsic architecture of neural networks
- altered intrinsic functional connectivity within networks linked to higher order cognition have been broadly linked to psychopathology across categorical disorders

Sample: 614 university students from the Duke Neurogenetics Study
- performed Connectome-Wide Association Study (CWAS) of the p factor
    - a data driven method for identifying resting-state seeds whose whole brain connectivity are associated with the p factor in the absence of a priori assumptions about brain regions or networks.
- 605/614 survived quality control: p factor ranged from 76.71to 191.96 with a mean of 99.80, sd of 15.39
- multi-dimensional matrix regression (MDMR) analysis revealed four regions with whole-brain connectivity patterns significantly associated with p factor scores:
    - left lingual gyrus
    - right middle occipital gyrus
    - two adjacent parcels of the left middle occipital gyrus
- the connectivity of each ROI
    - with visual and somatosensory regions decreased with increasing p factor scores
    - transmodal association regions increased with increasing p scores
- further analyses revealed the DMN and FPN as the major networks for which intrinsic functional connectivity was positively correlated with p factor scores
- a comparatively modest but notable negative correlation was observed between p factor scores and the intrinsic functional connectivity between the visual association cortex and somatomotor network
- **a circumscribed relationship between p factor scores and the whole-brain intrinsic connectivity of nodes in visual association cortex**
- hyper-connectivity suggests that increased risk for psychopathology may be manifest as greater effortful or less efficient executive control as well as poor regulation of self-referential information processing
